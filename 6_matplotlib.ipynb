```python
from matplotlib import pyplot as plt
import numpy as np
```

### basic plotting

```python
x = np.arange(1,10)
x
```

```python
y = np.random.randint(1,5,9)
y
```

```python
plt.plot(x,y)
plt.show()
```

### adding label

```python
plt.plot(x,y)
plt.title("Title for Graph")
plt.xlabel("Numbers")
plt.ylabel("Random Numbers")
plt.show()
```

### fromatting

```python
plt.plot(x,y, color="grey", linewidth=5, linestyle = ':')
plt.show()
```

```python
### plotting more than 1
x1 = np.arange(1,10)
x1
```

```python
y1 = x1*2
y1
```

```python
plt.plot(x,y, color="black", linewidth=5, linestyle = ':')
plt.plot(x1,y1, color="pink", linewidth=10)
plt.show()
```

### bar-plot *for categorical variable

```python
classroom = {'Aayush': 93, 'Abhaya': 85, 'Abhas': 82}
```

```python
name = list(classroom.keys())
marks = list(classroom.values())
```

```python
name,marks
```

```python
plt.bar(name, marks, color='black')
plt.title('Class Marks')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

```python
# for horizontal

plt.barh(name, marks, color='black') #added barh
plt.title('Class Marks')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

### scatter-plot *gives specific postion of x and y co-ordinates

```python
x = [4,8,12,3,4,5]
y = [1,2,3,7,8,9]
y1 = [1,2,3,4,5,6]
plt.scatter(x,y)
plt.scatter(x,y1, color="red")
```
